```python
Q1.
Ans:-
definition :- Abstraction in python is defined as a process of handling complexity by hiding unnecessary 
information from the user. This is one of the core concepts of object-oriented programming (OOP) languages. 
    
        # abstraction in python
from abc import ABC,abstractmethod          # (Abstract base class)

# abstract class
class Subject(ABC):
    @abstractmethod
    def subject(self):
        pass

class maths(Subject):
    def subject(self):
        print("Subject is Maths")

class Physics(Subject):
    def subject(self):
        print("Subject is Physics")

class Chemistry(Subject):
    def subject(self):
        print("Subject is Chemistry")

class English(Subject):
    def subject(self):
        print("Subject is English")

maths=Maths()
maths.subject()

physics=Physics()
physics.subject()

chemistry=Chemistry()
chemistry.subject()

english=English()
english.subject()

#Output:-
Subject is Maths
Subject is Physics
Subject is Chemistry
Subject is English
```

```python
Q2.
Ans:-
 Abstraction-  Abstraction is hiding the details and implementation of the code.
 Encapsulation - Encapsulation is hiding the (instance) data and controlling the visibility of the code.


#abstract method



import abc
from abc import ABC, abstractmethod

class Color(ABC):
    # an abstract method
    @abstractmethod
    def print_color(self):
        pass

class Red(Color):
    def print_color(self):
        print("The Color is red")

class Green(Color):
    def print_color(self):
        print("The Color is green")

class Blue(Color):
    def print_color(self):
        print("The Color is blue")

# trying to create an object of an abstract class
red = Red()
red.print_color()

green = Green()
green.print_color()

blue = Blue()
blue.print_color()

#ENCAPSULAITON

#encapsulaton method hiding instance used by __

    class Employee:
    # constructor
    def __init__(self, name, salary, project):           #__init__ hide instance
        # data members
        self.name = name
        self.salary = salary
        self.project = project
        
    def show(self):
        
        return "Name: ", self.name, 'Salary:', self.salary

 
    def work(self):
        return self.name, 'is working on', self.project

# creating object of a class
emp = Employee(input("enter Employee name"), 8000, 'NLP')

# calling public method of the class

emp.show()
emp.work()
```

```python
Q4.
Ans:-
Data abstraction is a programming concept that refers to the practice of hiding implementation details from
the user, and exposing only the relevant information or functionality. In Python, data abstraction can be 
achieved in several ways:

1. Using Classes: Classes provide a way to encapsulate data and behavior into a single object, and to hide the
   implementation details from the user. By defining a class with private data members and public methods that
   access or modify those data members, you can achieve data abstraction in Python.

2. Using Properties: Properties are a way to define methods that look like data attributes, but are actually
   computed dynamically. By using properties to encapsulate data, you can hide the implementation details
   from the user and provide a more abstract interface to the data.

3. Using Modules: Modules are a way to group related functions and data together into a single namespace. 
   By exposing only the relevant functions and data through the module's interface, you can achieve a 
    level of data abstraction in Python.

4. sing Abstract Base Classes: Abstract base classes are classes that define a set of methods that must be
   implemented by any concrete (i.e., non-abstract) subclasses. By using abstract base classes to define interfaces, you can 
   achieve a level of data abstraction in Python by hiding the implementation details of the concrete subclasses.

```
